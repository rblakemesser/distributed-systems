Goals of the course
==========================

Theory
----------------

The course relies heavily on theoretical / toy problems to demonstrate Distributed Systems concepts. For example:

### **BYZANTINE PROBLEM**
- Three generals have to attack an army that they surround
- They each have a binary decision to make (attack / retreat)
- One of them may be a traitor

The **BYZANTINE PROBLEM** is a theoretical problem in distributed systems. How can you get all nodes (generals) to agree on some action?

Assumptions of the course
-----------------------------

**DISTRIBUTED SYSTEMS** is taken to mean communiation over a network with multiple computers. We generally assume the primitives that each node can:

- Send message
- Receive message

And consider these to be desirable properties of a network:

- synchronization
- commonification

Three assumptions we make about Distributed Systems in the course are:

- **NO SHARED CLOCK** -- Time is relative to each node.
- **NO SHARED MEMORY** -- Each node must rely on local knowledge.
- **NO ACCURATE FAILURE DETECTION** -- Slow processes are indistinguishable from dead ones.

TCP and UDP
-------------------

Differences between TCP and UDP. As protocols, each represents a different strategy for sending messages.

Both assume:

- two computers on a network
- every computer on the network is addressed

TCP
----------------

- We can think of TCP as a pipe between sender and receiver
- Messages are received in the order they are sent
- Overhead associated with guaranteeing message integrity

UDP
----------------

- UDP can be thought of like sending a letter in the mail
- No individual packet is guaranteed to be received
- Less overhead because no receipt verification
