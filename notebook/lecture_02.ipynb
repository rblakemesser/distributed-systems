Set-theory notation and relation properties
===================
This lecture introduces a kind of discrete math analysis that allows us to describe the properties of a relation on a set.

A relation $R$ on a set $X$ can be written as:
> $$(X, R)$$

We define several properties that relations can have on sets...

Reflexive
-------------------

$$\forall_{x} \in X :$$
> $$( x, x ) \in R$$

The $R$ relation is reflexive on the set $X$ if for all possible $x$ in the set $X$, $(x, x) \in R$.

Irreflexive
-------------------

$$\forall_{x} \in X :$$
> $$( x, x ) \notin R$$

Irreflexive is the opposite. Not mutually exclusive though!

Symmetric
-------------------

$$ \forall_{x,y} \in X :$$
> $$( x, y ) \in R \implies (y, x) \in R$$

Symmetric always goes both ways

Asymmetric
-------------------

$$ \forall_{x,y} \in X :$$
> $$( x, y ) \in R \implies (y, x) \notin R$$

Asymmetric never goes both ways

Antisymmetric
-------------------

$$ \forall_{x,y} \in X :$$
> $$( x, y ) \in R \land (y, x) \in R \implies (x = y)$$


Transitive
-------------------

$$ \forall_{x,y,z} \in X:$$
> $$(x, y) \in R \land (y, z) \in R \implies (x, z) \in R$$

Enforced hierarchy. If $a > b$ and $b > c$ then $a > c$.

Examples!
-------------------

$(\mathbb{N}, \leq)$ The "less than or equal to" relation on the set of natural numbers:

> reflexive, antisymmetric, transitive

### Happened-before model ( $\to$ )

In set theory you can define whatever relation you want. We define the relation "Happened-before" to exist on a set of computational events.

The happened-before relation "$\to$" on a finite set of events "$E$" can be written as $( E , \to ):$

$e \to f$ when these conditions are met (using pseudo-math here):

- $\forall_{e,f} :$ if $e$ and $f$ are on the same process, and $e$ is executed before $f$ then $e \to f$
- $\forall_{e,f} :$ if $e$ is send of a message and $f$ is the corresponding receive event then $e \to f$
- $\exists_{g}: e \to g \land g \to f$ then $e \to f$

##### Concurrency ( $||$ )

Concurrency is the inverse of "happened-before". We define concurrency on the set of events in a distributed system $( E , || )$:

$ e || f $ when:

> $\lnot (e \to f) \land \lnot (f \to e)$

Simply put, if $e$ is concurrent with $f$, we don't know which happened first. Concurrency is indistinguishable from "happening at the same time".
